```python
import pandas as pd
import os

# 定义结果文件夹路径
results_dir = r'E:\myREPO\Fairness-Decision-Focused-Loss\Organized-FDFL\notebooks\results'
csv_files = [f for f in os.listdir(results_dir) if f.endswith('.csv')]
dataframes = {}
for csv_file in csv_files:
    file_path = os.path.join(results_dir, csv_file)
    df = pd.read_csv(file_path)
    dataframes[csv_file] = df
    print(f'Loaded {csv_file}: shape={df.shape}')
    display(df.head())
```

```python
import pandas as pd
import numpy as np

# 1. 定义一个基础的、通用的列名映射字典
# 这个字典处理在所有文件中命名一致或相似的列
base_rename_map = {
    # 统一性能指标
    'loss': 'Training Loss mean',
    'mse': 'Prediction MSE mean',
    'loss_std': 'Training Loss std',
    'mse_std': 'Prediction MSE std',
    
    'regret': 'Decision Regret mean',
    'regret_std': 'Decision Regret std',
    
    'fairness_std': 'Prediction Fairness std',
    'Fairness Penalty std': 'Prediction Fairness std',

    'G0 Decision Obj mean': 'G0 Objective mean',
    'G0 Objective': 'G0 Objective mean',
    'G0_decision_obj': 'G0 Objective mean',
    'G0 Decision Obj std': 'G0 Objective std',
    'G0_objective_std': 'G0 Objective std',
    'G0_decision_obj_std': 'G0 Objective std',

    'G1 Decision Obj mean': 'G1 Objective mean',
    'G1 Objective': 'G1 Objective mean',
    'G1_decision_obj': 'G1 Objective mean',
    'G1 Decision Obj std': 'G1 Objective std',
    'G1_objective_std': 'G1 Objective std',
    'G1_decision_obj_std': 'G1 Objective std',

    'G0 MSE mean': 'G0 MSE mean',
    'G0 MSE': 'G0 MSE mean',
    'G0_mse': 'G0 MSE mean',
    'G0 MSE std': 'G0 MSE std',
    'G0_mse_std': 'G0 MSE std',

    'G1 MSE mean': 'G1 MSE mean',
    'G1 MSE': 'G1 MSE mean',
    'G1_mse': 'G1 MSE mean',
    'G1 MSE std': 'G1 MSE std',
    'G1_mse_std': 'G1 MSE std',

    'G0 True Benefit mean': 'G0 True Benefit mean',
    'G0 Fairness': 'G0 True Benefit mean',
    'G0 True Benefit std': 'G0 True Benefit std',
    'G0_fairness_std': 'G0 True Benefit std',

    'G1 True Benefit mean': 'G1 True Benefit mean',
    'G1 Fairness': 'G1 True Benefit mean',
    'G1 True Benefit std': 'G1 True Benefit std',
    'G1_fairness_std': 'G1 True Benefit std',
    
    'training_time': 'Training Time mean',
    'training_time_std': 'Training Time std',
    
    'Predictor mean': 'Predictor'
}

processed_dfs = []

# 2. 遍历所有数据框进行处理
for name, df in dataframes.items():
    temp_df = df.copy()
    
    # 首先，应用通用的重命名规则
    temp_df.rename(columns=base_rename_map, inplace=True)
    
    # 接下来，根据文件来源处理特定的、命名冲突的列
    if 'pto' in name:
        # 在PTO文件中, 'fairness'是值, 'Fairness'是类型
        temp_df.rename(columns={'fairness': 'Prediction Fairness mean'}, inplace=True)
        # 'Fairness'列名已经是正确的类型标识，无需更改
        
    elif 'foldopt' in name:
        # 在FoldOpt文件中, 'Fairness Penalty mean'是值, 'FairnessType'是类型
        temp_df.rename(columns={
            'Fairness Penalty mean': 'Prediction Fairness mean',
            'FairnessType': 'Fairness'
        }, inplace=True)
        
    else: # for res-lr.csv and res-nn.csv
        # 在这些文件中, 'Prediction Fairness mean' 和 'Fairness' 列名已经是正确的
        pass

    # 添加来源标识列
    if 'foldopt' in name:
        temp_df['Method'] = 'FoldOpt'
    elif 'pto' in name:
        temp_df['Method'] = 'PTO'
    else: # for res-lr.csv and res-nn.csv
        # 这些文件包含closed-form和finite-diff方法，使用Grad Method列作为方法名
        if 'Grad Method' in temp_df.columns:
            temp_df['Method'] = temp_df['Grad Method']
        else:
            temp_df['Method'] = 'DFL'  # 默认标记，如果没有Grad Method列
        
    if 'nn' in name:
        temp_df['Predictor'] = 'NN'
    elif 'lr' in name:
        temp_df['Predictor'] = 'LR'
    
    # 删除不必要的列
    if 'Unnamed: 0' in temp_df.columns:
        temp_df.drop(columns=['Unnamed: 0'], inplace=True)
        
    processed_dfs.append(temp_df)

# 3. 合并所有数据框
master_df = pd.concat(processed_dfs, ignore_index=True, sort=False)

# 4. 显示合并后的结果以供验证
print("Master DataFrame Info:")
master_df.info()
print("\nColumns related to fairness in the final master_df:")
print(master_df[['Method', 'Predictor', 'Fairness', 'Prediction Fairness mean']].head())
print("\nMaster DataFrame Head:")
display(master_df.head())
print(f"\nTotal rows in Master DataFrame: {len(master_df)}")
```

```python
import numpy as np
import pandas as pd

def format_with_std(df, value_col, index, columns, aggfunc='mean'):
    """Helper function to create a pivot table with 'mean ± std' format."""
    mean_col = value_col
    std_col = value_col.replace('mean', 'std')

    if std_col not in df.columns:
        # If no std column, just show the mean
        mean_table = pd.pivot_table(df, values=mean_col, index=index, columns=columns, aggfunc=aggfunc)
        return mean_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-')

    mean_table = pd.pivot_table(df, values=mean_col, index=index, columns=columns, aggfunc=aggfunc)
    std_table = pd.pivot_table(df, values=std_col, index=index, columns=columns, aggfunc=aggfunc)

    # Combine mean and std into a single formatted string
    formatted_table = mean_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-') + \
                      " ± " + \
                      std_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-')
    
    # Clean up cases where data might be missing
    formatted_table = formatted_table.replace('- ± -', '-', regex=False)
    return formatted_table

def generate_tables(group_setting, fairness_type, alpha_values=[0.5, 2]):
    """
    Generates and displays a set of pivot tables for a specific group and fairness setting.
    
    Args:
        group_setting (bool): The group setting to filter by (True or False).
        fairness_type (str): The type of fairness to filter by ('mad' or 'atkinson').
        alpha_values (list): The list of Alpha values for columns.
    """
    # Determine lambda values based on fairness type
    if fairness_type == 'mad':
        lambda_values = [0, 0.05, 0.5]
    elif fairness_type == 'atkinson':
        lambda_values = [0, 1, 5]
    else:
        print(f"Error: Invalid fairness_type '{fairness_type}'. Please use 'mad' or 'atkinson'.")
        return

    print(f"======================================================================")
    print(f"    Displaying Results for: GROUP = {group_setting}, FAIRNESS = {fairness_type.upper()}")
    print(f"======================================================================")

    # 1. Filter master_df based on all specified criteria
    df = master_df[
        (master_df['Group'] == group_setting) &
        (master_df['Fairness'].str.contains(fairness_type, case=False, na=False)) &
        (master_df['Lambda'].isin(lambda_values)) &
        (master_df['Alpha'].isin(alpha_values))
    ].copy()

    if df.empty:
        print(f"No data found for the specified filters. Skipping tables.\n")
        return

    # 2. Create the 'Algorithm' column
    # 现在closed-form和finite-diff已经直接作为Method，不需要特殊处理
    df['Algorithm'] = df['Method']
    
    # 如果还有使用Two-Stage标记的数据，则使用Grad Method
    two_stage_mask = df['Method'] == 'Two-Stage'
    if two_stage_mask.any():
        df.loc[two_stage_mask, 'Algorithm'] = df.loc[two_stage_mask, 'Grad Method']
    
    # Define the new index
    table_index = ['Algorithm', 'Predictor', 'Fairness', 'Lambda']

    # --- Table 1: Decision Regret ---
    print("\n------ 1. Decision Performance: Decision Regret (mean ± std) ------")
    regret_table = format_with_std(df, 'Decision Regret mean', table_index, 'Alpha')
    display(regret_table)

    # --- Table 2: Prediction MSE ---
    print("\n------ 2. Prediction Performance: Prediction MSE (mean ± std) ------")
    n0, n1 = 4422, 578
    foldopt_mask = (df['Method'] == 'FoldOpt') & (df['Prediction MSE mean'].isna())
    if not df.loc[foldopt_mask].empty:
        # Calculate weighted mean
        g0_mse_mean = df.loc[foldopt_mask, 'G0 MSE mean']
        g1_mse_mean = df.loc[foldopt_mask, 'G1 MSE mean']
        df.loc[foldopt_mask, 'Prediction MSE mean'] = (g0_mse_mean * n0 + g1_mse_mean * n1) / (n0 + n1)
        # Calculate weighted std
        g0_mse_std = df.loc[foldopt_mask, 'G0 MSE std']
        g1_mse_std = df.loc[foldopt_mask, 'G1 MSE std']
        df.loc[foldopt_mask, 'Prediction MSE std'] = np.sqrt((g0_mse_std**2 * n0 + g1_mse_std**2 * n1) / (n0 + n1))

    mse_table = format_with_std(df, 'Prediction MSE mean', table_index, 'Alpha')
    display(mse_table)

    # --- Table 3: Prediction Fairness ---
    print(f"\n------ 3. Prediction Fairness: {fairness_type.upper()} (mean ± std) ------")
    fairness_table = format_with_std(df, 'Prediction Fairness mean', table_index, 'Alpha')
    display(fairness_table)

    # --- Table 4: Training Time ---
    print("\n------ 4. Runtime: Training Time (seconds, mean ± std) ------")
    time_table = format_with_std(df, 'Training Time mean', table_index, 'Alpha')
    display(time_table)
    print("\n\n")


# --- EXAMPLE USAGE: Uncomment the line for the tables you want to generate ---

# --- Main Results (MAD Fairness) ---
print("Generating tables for main paper (MAD Fairness)...")
generate_tables(group_setting=True, fairness_type='mad')
# generate_tables(group_setting=False, fairness_type='mad')

# --- Appendix Results (Atkinson Fairness) ---
# print("\nGenerating tables for Appendix (Atkinson Fairness)...")
# generate_tables(group_setting=True, fairness_type='atkinson')
# generate_tables(group_setting=False, fairness_type='atkinson')
```

```python
# 合并FoldOpt运行时间数据到现有的FoldOpt行中
import pandas as pd
import numpy as np

# 检查是否已经加载了FoldOpt运行时间数据
if 'times-foldopt.csv' in dataframes:
    foldopt_times_df = dataframes['times-foldopt.csv'].copy()
    
    print("Processing FoldOpt runtime data for merging...")
    print(f"Original FoldOpt times shape: {foldopt_times_df.shape}")
    print(f"FoldOpt times columns: {list(foldopt_times_df.columns)}")
    
    # 检查FoldOpt运行时间数据的结构
    print("\nFoldOpt runtime data preview:")
    display(foldopt_times_df.head())
    
    # 标准化FoldOpt运行时间数据的列名以便合并
    if 'FairnessType' in foldopt_times_df.columns:
        foldopt_times_df.rename(columns={'FairnessType': 'Fairness'}, inplace=True)
    
    # 确保数据类型匹配
    foldopt_times_df['Group'] = foldopt_times_df['Group'].astype(bool)
    
    # 重命名运行时间列为标准格式
    if 'Mean Runtime' in foldopt_times_df.columns:
        foldopt_times_df.rename(columns={
            'Mean Runtime': 'Training Time mean',
            'Std Runtime': 'Training Time std'
        }, inplace=True)
    
    # 检查主数据框中现有的FoldOpt数据
    print("\nChecking existing FoldOpt data in master_df...")
    foldopt_in_master = master_df[master_df['Method'] == 'FoldOpt'].copy()
    print(f"FoldOpt rows in master_df: {len(foldopt_in_master)}")
    
    if len(foldopt_in_master) > 0:
        print("Sample of existing FoldOpt data:")
        sample_cols = ['Method', 'Predictor', 'Group', 'Fairness', 'Alpha', 'Lambda', 'Training Time mean', 'Decision Regret mean']
        display(foldopt_in_master[sample_cols].head())
    
    # 准备合并的关键列
    merge_columns = ['Group', 'Fairness', 'Alpha', 'Lambda', 'Predictor']
    available_merge_cols = [col for col in merge_columns if col in foldopt_times_df.columns and col in master_df.columns]
    print(f"\nMerge columns available: {available_merge_cols}")
    
    # 准备运行时间数据用于合并（只选择需要的列）
    time_cols = ['Training Time mean', 'Training Time std']
    foldopt_runtime_merge = foldopt_times_df[available_merge_cols + time_cols].copy()
    
    print(f"\nRuntime data prepared for merge:")
    print(f"Shape: {foldopt_runtime_merge.shape}")
    display(foldopt_runtime_merge.head())
    
    # 备份原始数据
    master_df_backup = master_df.copy()
    
    # 找到所有FoldOpt行并更新运行时间数据
    foldopt_mask = master_df['Method'] == 'FoldOpt'
    foldopt_indices = master_df[foldopt_mask].index
    
    print(f"\nUpdating {len(foldopt_indices)} FoldOpt rows with runtime data...")
    
    updated_count = 0
    for idx in foldopt_indices:
        row = master_df.loc[idx]
        
        # 构建匹配条件来找到对应的运行时间数据
        match_conditions = []
        for col in available_merge_cols:
            match_conditions.append(foldopt_runtime_merge[col] == row[col])
        
        # 合并所有匹配条件
        final_condition = match_conditions[0]
        for condition in match_conditions[1:]:
            final_condition &= condition
        
        # 找到匹配的运行时间数据
        matching_times = foldopt_runtime_merge[final_condition]
        
        if len(matching_times) > 0:
            time_row = matching_times.iloc[0]
            
            # 只有当当前行缺少运行时间数据时才更新
            if pd.isna(master_df.loc[idx, 'Training Time mean']) or master_df.loc[idx, 'Training Time mean'] == 0:
                master_df.loc[idx, 'Training Time mean'] = time_row['Training Time mean']
                master_df.loc[idx, 'Training Time std'] = time_row['Training Time std']
                updated_count += 1
                if updated_count <= 10:  # 只显示前10个更新的示例
                    print(f"  Updated row {idx}: {row['Predictor']}-{row['Fairness']}-α{row['Alpha']}-λ{row['Lambda']} → "
                          f"Runtime = {time_row['Training Time mean']:.4f} ± {time_row['Training Time std']:.4f}")
            else:
                print(f"  Row {idx} already has runtime data, skipping")
        else:
            print(f"  No matching runtime data found for row {idx}: {row['Predictor']}-{row['Fairness']}-α{row['Alpha']}-λ{row['Lambda']}")
    
    print(f"\nSummary: Updated runtime data for {updated_count} FoldOpt rows")
    
    # 验证更新结果
    print("\n" + "="*70)
    print("VERIFICATION OF FOLDOPT RUNTIME DATA UPDATE")
    print("="*70)
    
    updated_foldopt = master_df[master_df['Method'] == 'FoldOpt'].copy()
    
    # 统计运行时间数据的完整性
    total_foldopt = len(updated_foldopt)
    has_runtime = (~updated_foldopt['Training Time mean'].isna()).sum()
    missing_runtime = (updated_foldopt['Training Time mean'].isna()).sum()
    
    print(f"Total FoldOpt rows: {total_foldopt}")
    print(f"Rows with runtime data: {has_runtime}")
    print(f"Rows missing runtime data: {missing_runtime}")
    print(f"Runtime data coverage: {has_runtime/total_foldopt*100:.1f}%")
    
    if has_runtime > 0:
        print(f"\nRuntime Statistics for FoldOpt:")
        time_stats = updated_foldopt[['Training Time mean', 'Training Time std']].describe()
        display(time_stats)
        
        print(f"\nSample of updated FoldOpt data with complete information:")
        sample_cols = ['Method', 'Predictor', 'Group', 'Fairness', 'Alpha', 'Lambda', 
                      'Decision Regret mean', 'Prediction MSE mean', 'Training Time mean', 'Training Time std']
        display(updated_foldopt[sample_cols].head(10))
    
    if missing_runtime > 0:
        print(f"\nRows still missing runtime data:")
        missing_rows = updated_foldopt[updated_foldopt['Training Time mean'].isna()]
        missing_summary = missing_rows[['Predictor', 'Group', 'Fairness', 'Alpha', 'Lambda']].drop_duplicates()
        print(f"Missing runtime data for {len(missing_summary)} unique parameter combinations:")
        display(missing_summary)
    
    print(f"\n✅ FoldOpt runtime data integration completed!")
    print(f"Successfully filled {updated_count} existing FoldOpt rows with runtime data")
    
else:
    print("Error: FoldOpt runtime data not found in dataframes dictionary.")
    print("Please make sure to run the previous cell to load times-foldopt.csv first.")
```

# 📊 **CLEAN RESULTS PROCESSING PIPELINE**

This notebook provides a streamlined pipeline for processing and analyzing experimental results from fairness algorithm comparisons.

## **Pipeline Overview:**
1. **Data Loading** - Import all CSV result files
2. **Data Standardization** - Unify column names and data types
3. **Data Integration** - Merge runtime data and clean duplicates
4. **Table Generation** - Create publication-ready results tables

```python
# ==============================================================================
# 🔧 CONFIGURATION AND IMPORTS
# ==============================================================================
import pandas as pd
import numpy as np
import os
from typing import Dict, List

# Configuration
RESULTS_DIR = r'E:\myREPO\Fairness-Decision-Focused-Loss\Organized-FDFL\notebooks\results'
RUNTIME_FILE = 'times-foldopt.csv'

print("📦 Imports completed successfully!")
```

```python
# ==============================================================================
# 🏗️ DATA PROCESSING CLASS
# ==============================================================================

class ResultsProcessor:
    """Comprehensive results processing pipeline for fairness algorithm experiments."""
    
    def __init__(self, results_dir: str):
        self.results_dir = results_dir
        self.dataframes = {}
        self.master_df = None
        
        # Column standardization mapping
        self.base_rename_map = {
            'loss': 'Training Loss mean', 'mse': 'Prediction MSE mean',
            'loss_std': 'Training Loss std', 'mse_std': 'Prediction MSE std',
            'regret': 'Decision Regret mean', 'regret_std': 'Decision Regret std',
            'fairness_std': 'Prediction Fairness std',
            'Fairness Penalty std': 'Prediction Fairness std',
            'training_time': 'Training Time mean', 'training_time_std': 'Training Time std',
            'Predictor mean': 'Predictor',
            
            # G0/G1 Objective columns
            'G0 Decision Obj mean': 'G0 Objective mean', 'G0 Objective': 'G0 Objective mean',
            'G0_decision_obj': 'G0 Objective mean', 'G0 Decision Obj std': 'G0 Objective std',
            'G0_objective_std': 'G0 Objective std', 'G0_decision_obj_std': 'G0 Objective std',
            'G1 Decision Obj mean': 'G1 Objective mean', 'G1 Objective': 'G1 Objective mean',
            'G1_decision_obj': 'G1 Objective mean', 'G1 Decision Obj std': 'G1 Objective std',
            'G1_objective_std': 'G1 Objective std', 'G1_decision_obj_std': 'G1 Objective std',
            
            # G0/G1 MSE columns
            'G0 MSE mean': 'G0 MSE mean', 'G0 MSE': 'G0 MSE mean', 'G0_mse': 'G0 MSE mean',
            'G0 MSE std': 'G0 MSE std', 'G0_mse_std': 'G0 MSE std',
            'G1 MSE mean': 'G1 MSE mean', 'G1 MSE': 'G1 MSE mean', 'G1_mse': 'G1 MSE mean',
            'G1 MSE std': 'G1 MSE std', 'G1_mse_std': 'G1 MSE std',
            
            # G0/G1 Fairness columns
            'G0 True Benefit mean': 'G0 True Benefit mean', 'G0 Fairness': 'G0 True Benefit mean',
            'G0 True Benefit std': 'G0 True Benefit std', 'G0_fairness_std': 'G0 True Benefit std',
            'G1 True Benefit mean': 'G1 True Benefit mean', 'G1 Fairness': 'G1 True Benefit mean',
            'G1 True Benefit std': 'G1 True Benefit std', 'G1_fairness_std': 'G1 True Benefit std'
        }
    
    def load_all_csvs(self) -> Dict[str, pd.DataFrame]:
        """Load all CSV files from the results directory."""
        print("📂 Loading CSV files...")
        
        csv_files = [f for f in os.listdir(self.results_dir) if f.endswith('.csv')]
        
        for csv_file in csv_files:
            file_path = os.path.join(self.results_dir, csv_file)
            df = pd.read_csv(file_path)
            self.dataframes[csv_file] = df
            print(f"   ✓ {csv_file}: {df.shape}")
        
        print(f"📊 Loaded {len(csv_files)} CSV files")
        return self.dataframes
    
    def standardize_dataframe(self, df: pd.DataFrame, filename: str) -> pd.DataFrame:
        """Standardize column names and add method/predictor identifiers."""
        temp_df = df.copy()
        
        # Apply base rename mapping
        temp_df.rename(columns=self.base_rename_map, inplace=True)
        
        # File-specific column handling
        if 'pto' in filename:
            temp_df.rename(columns={'fairness': 'Prediction Fairness mean'}, inplace=True)
            temp_df['Method'] = 'PTO'
        elif 'foldopt' in filename:
            temp_df.rename(columns={
                'Fairness Penalty mean': 'Prediction Fairness mean',
                'FairnessType': 'Fairness'
            }, inplace=True)
            temp_df['Method'] = 'FoldOpt'
        else:  # res-lr.csv and res-nn.csv
            if 'Grad Method' in temp_df.columns:
                temp_df['Method'] = temp_df['Grad Method']
            else:
                temp_df['Method'] = 'DFL'
        
        # Add predictor information
        if 'nn' in filename:
            temp_df['Predictor'] = 'NN'
        elif 'lr' in filename:
            temp_df['Predictor'] = 'LR'
        
        # Clean up
        if 'Unnamed: 0' in temp_df.columns:
            temp_df.drop(columns=['Unnamed: 0'], inplace=True)
        
        return temp_df
    
    def create_master_dataframe(self) -> pd.DataFrame:
        """Create unified master dataframe from all loaded files."""
        print("🔄 Creating master dataframe...")
        
        processed_dfs = []
        for filename, df in self.dataframes.items():
            if 'times-' not in filename:  # Skip runtime files for now
                standardized_df = self.standardize_dataframe(df, filename)
                processed_dfs.append(standardized_df)
                print(f"   ✓ Processed {filename}")
        
        self.master_df = pd.concat(processed_dfs, ignore_index=True, sort=False)
        print(f"📊 Master dataframe created: {self.master_df.shape}")
        return self.master_df
    
    def integrate_runtime_data(self, runtime_filename: str = 'times-foldopt.csv'):
        """Integrate FoldOpt runtime data into master dataframe."""
        if runtime_filename not in self.dataframes:
            print(f"⚠️  Runtime file {runtime_filename} not found")
            return
        
        print("⏱️  Integrating runtime data...")
        runtime_df = self.dataframes[runtime_filename].copy()
        
        # Standardize runtime data
        if 'FairnessType' in runtime_df.columns:
            runtime_df.rename(columns={'FairnessType': 'Fairness'}, inplace=True)
        if 'Mean Runtime' in runtime_df.columns:
            runtime_df.rename(columns={
                'Mean Runtime': 'Training Time mean',
                'Std Runtime': 'Training Time std'
            }, inplace=True)
        
        # Fix predictor naming in runtime data
        runtime_df['Predictor'] = runtime_df['Predictor'].replace({
            'LinearRegressionModel': 'LR', 
            'FairRiskPredictor': 'NN'
        })
        
        runtime_df['Group'] = runtime_df['Group'].astype(bool)
        
        # Merge runtime data directly into master_df for FoldOpt rows
        merge_cols = ['Group', 'Fairness', 'Alpha', 'Lambda', 'Predictor']
        runtime_cols = ['Training Time mean', 'Training Time std']
        
        # Create a subset of runtime data for merging
        runtime_for_merge = runtime_df[merge_cols + runtime_cols].copy()
        
        # Update FoldOpt rows with runtime data
        foldopt_mask = self.master_df['Method'] == 'FoldOpt'
        updated_count = 0
        
        for idx in self.master_df[foldopt_mask].index:
            row = self.master_df.loc[idx]
            
            # Find matching runtime data
            match_mask = (
                (runtime_for_merge['Group'] == row['Group']) &
                (runtime_for_merge['Fairness'] == row['Fairness']) &
                (runtime_for_merge['Alpha'] == row['Alpha']) &
                (runtime_for_merge['Lambda'] == row['Lambda']) &
                (runtime_for_merge['Predictor'] == row['Predictor'])
            )
            
            matching_runtime = runtime_for_merge[match_mask]
            
            if len(matching_runtime) > 0:
                time_data = matching_runtime.iloc[0]
                self.master_df.loc[idx, 'Training Time mean'] = time_data['Training Time mean']
                self.master_df.loc[idx, 'Training Time std'] = time_data['Training Time std']
                updated_count += 1
        
        print(f"   ✓ Updated {updated_count} FoldOpt rows with runtime data")
        print(f"📊 Final master dataframe: {self.master_df.shape}")
        
    def get_summary(self) -> Dict[str, pd.Series]:
        """Get summary statistics of the master dataframe."""
        summary = {
            'Has_Results': self.master_df.groupby('Method').agg({
                'Decision Regret mean': lambda x: (~pd.isna(x)).sum()
            })['Decision Regret mean'],
            'Has_Runtime': self.master_df.groupby('Method').agg({
                'Training Time mean': lambda x: (~pd.isna(x)).sum()
            })['Training Time mean']
        }
        return summary

print("🏗️  ResultsProcessor class defined successfully!")
```

```python
# ==============================================================================
# 📊 TABLE GENERATION UTILITIES
# ==============================================================================

class TableGenerator:
    """Generate publication-ready tables from processed results."""
    
    def __init__(self, master_df: pd.DataFrame):
        self.master_df = master_df
    
    def format_with_std(self, df: pd.DataFrame, value_col: str, index: List[str], 
                       columns: str, aggfunc='mean') -> pd.DataFrame:
        """Create pivot table with 'mean ± std' format."""
        mean_col = value_col
        std_col = value_col.replace('mean', 'std')
        
        if std_col not in df.columns:
            mean_table = pd.pivot_table(df, values=mean_col, index=index, 
                                      columns=columns, aggfunc=aggfunc)
            return mean_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-')
        
        mean_table = pd.pivot_table(df, values=mean_col, index=index, 
                                  columns=columns, aggfunc=aggfunc)
        std_table = pd.pivot_table(df, values=std_col, index=index, 
                                 columns=columns, aggfunc=aggfunc)
        
        formatted_table = mean_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-') + \
                         " ± " + \
                         std_table.map(lambda x: f'{x:.4f}' if pd.notna(x) else '-')
        
        return formatted_table.replace('- ± -', '-', regex=False)
    
    def to_latex(self, table: pd.DataFrame, table_name: str, group_setting: bool, fairness_type: str) -> str:
        """Convert pivot table to LaTeX format."""
        
        # Create caption and label
        group_str = "with" if group_setting else "without"
        caption = f"{table_name} results ({'Group' if group_str else 'Individual'} Alpha Decision, {fairness_type.upper()} Prediction Fairness)"
        label = f"tab:{table_name.lower().replace(' ', '_')}_{fairness_type}_{group_str}_group"
        
        # Start LaTeX table
        latex_lines = [
            "\\begin{table}[htbp]",
            "\\centering",
            f"\\caption{{{caption}}}",
            f"\\label{{{label}}}",
            "\\footnotesize",  # Make font smaller for better fit
        ]
        
        # Calculate number of columns (index levels + data columns)
        n_index_cols = len(table.index.names) if hasattr(table.index, 'names') else 1
        n_data_cols = len(table.columns)
        total_cols = n_index_cols + n_data_cols
        
        # Column specification
        col_spec = "l" * n_index_cols + "c" * n_data_cols
        latex_lines.append(f"\\begin{{tabular}}{{{col_spec}}}")
        latex_lines.append("\\toprule")
        
        # Header row
        index_names = []
        if hasattr(table.index, 'names'):
            index_names = [name if name else "" for name in table.index.names]
        else:
            index_names = [""]
        
        # Create header
        header_parts = index_names + [f"$\\alpha = {col}$" for col in table.columns]
        header_line = " & ".join(header_parts) + " \\\\"
        latex_lines.append(header_line)
        latex_lines.append("\\midrule")
        
        # Data rows
        for idx, row in table.iterrows():
            # Handle multi-index
            if isinstance(idx, tuple):
                row_parts = list(idx)
            else:
                row_parts = [str(idx)]
            
            # Format algorithm names for LaTeX
            for i, part in enumerate(row_parts):
                if part == 'closed-form':
                    row_parts[i] = 'DFL-CF'
                elif part == 'finite-diff':
                    row_parts[i] = 'DFL-FD'
                elif part == 'FoldOpt':
                    row_parts[i] = 'FoldOpt'
                elif part == 'PTO':
                    row_parts[i] = 'PTO'
                else:
                    row_parts[i] = str(part)
            
            # Add data values
            data_parts = [str(val) if pd.notna(val) else "-" for val in row.values]
            
            # Combine all parts
            all_parts = row_parts + data_parts
            row_line = " & ".join(all_parts) + " \\\\"
            latex_lines.append(row_line)
        
        # End table
        latex_lines.extend([
            "\\bottomrule",
            "\\end{tabular}",
            "\\end{table}",
            ""
        ])
        
        return "\n".join(latex_lines)
    
    def generate_all_latex_tables(self, alpha_values: List[float] = [0.5, 2]):
        """Generate LaTeX code for all table configurations."""
        
        fairness_configs = [
            ('mad', True),
            ('mad', False), 
            ('atkinson', True),
            ('atkinson', False)
        ]
        
        print("🎯 Generating LaTeX code for all publication-ready tables...")
        print("="*100)
        
        all_latex = []
        
        for fairness_type, group_setting in fairness_configs:
            print(f"\n📋 Processing: {fairness_type.upper()} fairness, Group = {group_setting}")
            
            # Generate tables for this configuration
            latex_tables = self._generate_latex_for_config(group_setting, fairness_type, alpha_values)
            all_latex.extend(latex_tables)
        
        # Save all LaTeX to a single string
        complete_latex = "\n\n".join(all_latex)
        
        print(f"\n✅ Generated LaTeX for {len(all_latex)} tables")
        print("\n" + "="*60)
        print("COMPLETE LATEX CODE:")
        print("="*60)
        print(complete_latex)
        
        return complete_latex
    
    def _generate_latex_for_config(self, group_setting: bool, fairness_type: str, 
                                  alpha_values: List[float]) -> List[str]:
        """Generate LaTeX tables for a specific configuration."""
        
        # Determine lambda values
        lambda_values = {
            'mad': [0, 0.05, 0.5],
            'atkinson': [0, 1, 5]
        }.get(fairness_type, [0])
        
        # Filter data
        df = self.master_df[
            (self.master_df['Group'] == group_setting) &
            (self.master_df['Fairness'].str.contains(fairness_type, case=False, na=False)) &
            (self.master_df['Lambda'].isin(lambda_values)) &
            (self.master_df['Alpha'].isin(alpha_values))
        ].copy()
        
        if df.empty:
            return []
        
        # Create Algorithm column
        df['Algorithm'] = df['Method']
        two_stage_mask = df['Method'] == 'Two-Stage'
        if two_stage_mask.any():
            df.loc[two_stage_mask, 'Algorithm'] = df.loc[two_stage_mask, 'Grad Method']
        
        table_index = ['Algorithm', 'Predictor', 'Fairness', 'Lambda']
        
        # Generate tables
        tables = {
            'Decision Regret': 'Decision Regret mean',
            'Prediction MSE': 'Prediction MSE mean', 
            'Prediction Fairness': 'Prediction Fairness mean',
            'Training Time': 'Training Time mean'
        }
        
        latex_tables = []
        
        for table_name, metric_col in tables.items():
            # Special handling for FoldOpt MSE calculation
            if metric_col == 'Prediction MSE mean':
                n0, n1 = 4422, 578
                foldopt_mask = (df['Method'] == 'FoldOpt') & (df[metric_col].isna())
                if not df.loc[foldopt_mask].empty:
                    g0_mse_mean = df.loc[foldopt_mask, 'G0 MSE mean']
                    g1_mse_mean = df.loc[foldopt_mask, 'G1 MSE mean']
                    df.loc[foldopt_mask, metric_col] = (g0_mse_mean * n0 + g1_mse_mean * n1) / (n0 + n1)
                    
                    g0_mse_std = df.loc[foldopt_mask, 'G0 MSE std']
                    g1_mse_std = df.loc[foldopt_mask, 'G1 MSE std']
                    df.loc[foldopt_mask, 'Prediction MSE std'] = np.sqrt((g0_mse_std**2 * n0 + g1_mse_std**2 * n1) / (n0 + n1))
            
            table = self.format_with_std(df, metric_col, table_index, 'Alpha')
            latex_code = self.to_latex(table, table_name, group_setting, fairness_type)
            latex_tables.append(latex_code)
        
        return latex_tables
    
    def generate_tables(self, group_setting: bool, fairness_type: str, 
                       alpha_values: List[float] = [0.5, 2], output_latex: bool = False):
        """Generate complete set of results tables."""
        
        # Determine lambda values
        lambda_values = {
            'mad': [0, 0.05, 0.5],
            'atkinson': [0, 1, 5]
        }.get(fairness_type, [0])
        
        print(f"{'='*80}")
        print(f"    📋 RESULTS FOR: GROUP = {group_setting}, FAIRNESS = {fairness_type.upper()}")
        print(f"{'='*80}")
        
        # Filter data
        df = self.master_df[
            (self.master_df['Group'] == group_setting) &
            (self.master_df['Fairness'].str.contains(fairness_type, case=False, na=False)) &
            (self.master_df['Lambda'].isin(lambda_values)) &
            (self.master_df['Alpha'].isin(alpha_values))
        ].copy()
        
        if df.empty:
            print(f"⚠️  No data found for specified filters")
            return
        
        # Create Algorithm column
        df['Algorithm'] = df['Method']
        two_stage_mask = df['Method'] == 'Two-Stage'
        if two_stage_mask.any():
            df.loc[two_stage_mask, 'Algorithm'] = df.loc[two_stage_mask, 'Grad Method']
        
        table_index = ['Algorithm', 'Predictor', 'Fairness', 'Lambda']
        
        # Generate tables
        tables = {
            'Decision Regret': 'Decision Regret mean',
            'Prediction MSE': 'Prediction MSE mean', 
            'Prediction Fairness': 'Prediction Fairness mean',
            'Training Time': 'Training Time mean'
        }
        
        for i, (table_name, metric_col) in enumerate(tables.items(), 1):
            print(f"\n------ {i}. {table_name} (mean ± std) ------")
            
            # Special handling for FoldOpt MSE calculation
            if metric_col == 'Prediction MSE mean':
                n0, n1 = 4422, 578
                foldopt_mask = (df['Method'] == 'FoldOpt') & (df[metric_col].isna())
                if not df.loc[foldopt_mask].empty:
                    g0_mse_mean = df.loc[foldopt_mask, 'G0 MSE mean']
                    g1_mse_mean = df.loc[foldopt_mask, 'G1 MSE mean']
                    df.loc[foldopt_mask, metric_col] = (g0_mse_mean * n0 + g1_mse_mean * n1) / (n0 + n1)
                    
                    g0_mse_std = df.loc[foldopt_mask, 'G0 MSE std']
                    g1_mse_std = df.loc[foldopt_mask, 'G1 MSE std']
                    df.loc[foldopt_mask, 'Prediction MSE std'] = np.sqrt((g0_mse_std**2 * n0 + g1_mse_std**2 * n1) / (n0 + n1))
            
            table = self.format_with_std(df, metric_col, table_index, 'Alpha')
            
            if output_latex:
                latex_code = self.to_latex(table, table_name, group_setting, fairness_type)
                print(f"\nLaTeX code for {table_name}:")
                print(latex_code)
            else:
                display(table)
        
        print(f"\n{'='*80}")

print("📊 TableGenerator class defined successfully!")
```

```python
# ==============================================================================
# 🚀 EXECUTION PIPELINE 
# ==============================================================================

def create_complete_results_pipeline():
    """Complete pipeline to process all experimental results."""
    
    print("🔄 Starting complete results processing pipeline...")
    print("="*80)
    
    # Initialize processor with results directory
    processor = ResultsProcessor(RESULTS_DIR)
    
    # Load and process all data
    print("📂 Loading all CSV files...")
    csv_files = processor.load_all_csvs()
    print(f"✅ Loaded {len(csv_files)} files: {list(csv_files.keys())}")
    
    # Create master dataframe
    print("\n🔗 Creating master dataframe...")
    master_df = processor.create_master_dataframe()
    print(f"✅ Master dataframe created with shape: {master_df.shape}")
    
    # Integrate runtime data
    print("\n⏱️  Integrating runtime data...")
    processor.integrate_runtime_data(RUNTIME_FILE)
    master_df = processor.master_df  # Update with integrated data
    
    # Show summary
    print("\n📊 Data Summary:")
    summary = processor.get_summary()
    for key, value in summary.items():
        print(f"  • {key}: {value}")
    
    # Initialize table generator
    print("\n📋 Initializing table generator...")
    table_gen = TableGenerator(master_df)
    
    return processor, table_gen, master_df

# Run the complete pipeline
processor, table_gen, master_df = create_complete_results_pipeline()
```
